```python
import matplotlib.pyplot as plt
import random
```

```python
def random_scale(numbers, min_scale=0.5, max_scale=1.5):
    return [x * random.uniform(min_scale, max_scale) for x in numbers]


x =  list(range(0,10))
y = random_scale(x)

print(x)
print(y)
plt.plot(x,y,marker="*",color="green")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = ["red","blue","green","purple"]
y = [1,5,32,14]

plt.bar(x,y,color="orange")
plt.title("Example")
plt.show()
```

```python
x = ["red","blue","green","orange"]
y = [1,5,32,14]

plt.pie(y,labels=x,autopct="%1.1f%%",colors=x)
plt.title("Example")
plt.show()
```
